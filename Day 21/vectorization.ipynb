```python
import numpy as np
import time

x=np.random.rand(100)
y=np.random.rand(100)
t=time.time()
#Vectorization
z=np.dot(x,y)
duration=time.time()-t
print(f'z ={z}-- {duration*1000}')

z=0
t=time.time()
# normal looping
for i in range(100):
  z+=x[i]*y[i]

duration=time.time()-t
print(f'z ={z}-- {duration*1000}')
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
t=time.time()
x=x-0.1*y
duration=time.time()-t
print(f'z ={x}-- {duration*1000}')


t=time.time()

for i in range(100):
  x+=x[i]*y[i]

duration=time.time()-t
print(f'z ={x}-- {duration*1000}')
```
